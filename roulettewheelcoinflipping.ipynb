```python
import random
import matplotlib.pyplot as plt


def spin_roulette_wheel(outcomes):
    # Calculate the total weight of all outcomes
    total_weight = sum(outcomes.values())

    # Generate a random number between 0 and total_weight
    rand_num = random.uniform(0, total_weight)

    # Iterate through the outcomes and find the winning outcome
    cumulative_weight = 0
    for outcome, weight in outcomes.items():
        cumulative_weight += weight
        if rand_num <= cumulative_weight:
            return outcome

def run_experiment(probability, num_trials):
    # Define the outcomes for the coin flip (Heads and Tails)
    outcomes = {
        "Heads": probability,
        "Tails": 1 - probability
    }

    # Run the experiment
    results = [spin_roulette_wheel(outcomes) for _ in range(num_trials)]
    return results

def plot_results(results):
    # Count the frequency of each outcome
    counts = {outcome: results.count(outcome) for outcome in set(results)}

    # Plot results as a pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(counts.values(), labels=counts.keys(), autopct='%1.1f%%')
    plt.title('Coin Flipping Experiment')
    plt.show()


# Parameters
probability = float(input("Enter the probability of getting Heads (between 0 and 1): "))
num_trials = int(input("Enter the number of trials: "))

# Run experiment
results = run_experiment(probability, num_trials)

# Analyze results and print
heads_count = results.count("Heads")
tails_count = results.count("Tails")
print(f"Head: {heads_count} times, Tail: {tails_count} times")
plot_results(results)
```
